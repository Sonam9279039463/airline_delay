```python
# Load dataset
data = pd.read_csv(r"C:\Users\Admin\Downloads\airline_delay.csv")

# View first few rows
print(data.head())
```

```python
print(data.tail())
```

```python
plt.figure(figsize=(10,6))

# Select only numeric columns
numeric_data = data.select_dtypes(include=['number'])

# Generate heatmap
sns.heatmap(numeric_data.corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Heatmap - Correlation Matrix')
plt.show()

```

```python
plt.figure(figsize=(12,6))
sns.lineplot(x=data['year'], y=data['arr_delay'], hue=data['carrier_name'])
plt.title('Flight Delays Over Time')
plt.xlabel('Year')
plt.ylabel('Average Arrival Delay')
plt.show()

```

```python
delay_reasons = ['carrier_delay', 'weather_delay', 'nas_delay', 'security_delay', 'late_aircraft_delay']
data[delay_reasons].sum().plot(kind='bar', color=['blue', 'red', 'green', 'purple', 'orange'])
plt.title('Total Delay Reasons')
plt.ylabel('Delay Minutes')
plt.show()
```

```python
plt.figure(figsize=(14,6))
sns.boxplot(x='carrier_name', y='arr_delay', data=data)
plt.xticks(rotation=90)
plt.title('Airline-Wise Delay Analysis')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='month', y='arr_delay', data=data)
plt.title('Monthly Variation in Flight Delays')
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.plot(data['year'], data['arr_delay'], marker='o', linestyle='-', color='b')
plt.xlabel('Year')
plt.ylabel('Average Arrival Delay')
plt.title('Line Plot - Flight Delays Over Time')
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.scatter(data['carrier_ct'], data['arr_delay'], color='r', alpha=0.6)
plt.xlabel('Carrier Delay')
plt.ylabel('Arrival Delay')
plt.title('Scatter Plot')
plt.show()

# Seaborn Version
sns.scatterplot(x=data['carrier_ct'], y=data['arr_delay'], hue=data['carrier_name'])
plt.title('Seaborn Scatter Plot')
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.hist(data['arr_delay'], bins=30, color='g', edgecolor='black')
plt.xlabel('Arrival Delay')
plt.ylabel('Frequency')
plt.title('Histogram - Distribution of Delays')
plt.show()

# Seaborn Version
sns.histplot(data['arr_delay'], bins=30, kde=True)
plt.title('Seaborn Histogram')
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.boxplot(data['arr_delay'])
plt.title('Box Plot - Flight Delays')
plt.show()

# Seaborn Version
sns.boxplot(y=data['arr_delay'])
plt.title('Seaborn Box Plot')
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.bar(data['carrier_name'], data['arr_delay'], color='orange')
plt.xticks(rotation=90)
plt.xlabel('Airline')
plt.ylabel('Total Delay Minutes')
plt.title('Bar Chart - Airline-wise Delays')
plt.show()

# Seaborn Version
sns.barplot(x=data['carrier_name'], y=data['arr_delay'])
plt.xticks(rotation=90)
plt.title('Seaborn Bar Chart')
plt.show()
```
